```python
### FETCH FROM DATABASE

# Connect to the database
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             database='covid19',
                             cursorclass=pymysql.cursors.DictCursor)

with connection:
    with connection.cursor() as cursor:
        # Create a new record
        sql = "select * from pandemic where cases > 10"
        cursor.execute(sql)

        result = cursor.fetchall()
#         print(result)
        
df = pd.DataFrame(result)
df_no_global = df[df.country != "Global"].sort_values(by = "cases").head(50)
```
